```
#teste de uma lista ordenada de forma decrescente
import time

def bubbleSort(lista):
    for num in range(len(lista)-1,0,-1):
        for i in range(num):
            if lista[i] > lista[i+1]:
                lista[i], lista[i+1] = lista[i+1], lista[i]

def selectionSort(lista):
    for posicao in range(len(lista)):
        posicao_menor = posicao
        for x in range(posicao+1, len(lista)):
            if lista[posicao_menor] > lista[x]:
                posicao_menor = x
        lista[posicao], lista[posicao_menor] = lista[posicao_menor], lista[posicao]

def insertionSort(lista):
    for posicao in range(1, len(lista)):
        key = lista[posicao]
        y = posicao-1
        while y >= 0 and key < lista[y] :
                lista[y+1] = lista[y]
                y -= 1
        lista[y+1] = key

lista = [i for i in range(200, 0, -1)]
print(f"a lista antes de ser ordenada:\n {lista}\n")

#BubbleSort
timerComecoB = time.time()
listaB = [i for i in range(200, 0, -1)]
bubbleSort(listaB)
timerFinalB = time.time() - timerComecoB
print(f"Bubblesort:\nO tempo desse sort foi {timerFinalB} segundos\nA lista apos ser ordenada:\n{listaB}\n")

#SelectionSort
timerComecoS = time.time()
listaS = [i for i in range(200, 0, -1)]
selectionSort(listaS)
timerFinalS = time.time() - timerComecoS
print(f"SelectionSort:\nO tempo desse sort foi {timerFinalS} segundos\nA lista apos ser ordenada:\n{listaS}\n")

#InsertionSort
timerComecoI = time.time()
listaI = [i for i in range(200, 0, -1)]
insertionSort(listaI)
timerFinalI = time.time() - timerComecoI
print(f"InserctionSort:\nO tempo desse sort foi {timerFinalI} segundos\nA lista apos ser ordenada:\n{listaI}\n")
```
