```
#programa que lê e transcreve ips
ips = ['192.168.0.256\n',
         '200.135.80.9\n',
         '192.168.1.1\n',
         '8.35.67.74\n',
         '257.32.4.5\n',
         '85.345.1.2\n',
         '9.8.234.5\n',
         '1.2.3.4\n'
         ]
ips_validos = ['200.135.80.9\n',
               '192.168.1.1\n',
               '8.35.67.74\n',
               '1.2.3.4\n'
               ]

#coloca os ips para verificar em um arquivo
arquivo1 = open("verificar.txt", "w+")
print("os ips para verificar são:\n--------------------------")
for x in ips:
  arquivo1.writelines(x)
  print(x)
arquivo1.close()

#o arquivo para verificar é aberto em leitura por já ter sido escrito, os outros dois são abertos em escrita para serem usados mais para frente
arquivo1 = open("verificar.txt", "r")
arquivo2 = open("ips_validos.txt", "w+")
arquivo3 = open("ips_verificados.txt", "w+")

#coloca os ips validos em outro arquivo
print("-------------------\nos ips validos são:\n-------------------")
for x in ips_validos:
  print(x)
  arquivo2.writelines(x)
arquivo2.close()
arquivo2 = open("ips_validos.txt", "r")

verificar = set(arquivo1).intersection(arquivo2)# essa variavel compara a lista de ips para verificar com os validos e coloca em um "set"
verificar.discard('\n')# isso tira os espaços das listas para ser mais facil de verificar

#aqui é escrito os arquivos comparados em um terceiro arquivo
for x in verificar:
  arquivo3.writelines(x)

#aqui imprime o  arquivo de ips comparados
print("---------------------------------------------------\nlogo, os ips validos no arquivo para verificar são:\n---------------------------------------------------")
arquivo3.close()
arquivo3 = open("ips_verificados.txt", "r")
for x in arquivo3.readlines():
  print(x)

arquivo1.close()
arquivo2.close()
arquivo3.close()
```
